## Import the pyplot module

`matplotlib` has many different modules, we'll be using `pyplot` so let's import that and give it an alias as we'll need to use dot notation to use function

```python
import matplotlib.pyplot as plt
```

## Sample Data

```python
sunny_days = [8, 10, 7, 14, 20, 18, 25, 19, 18, 14, 12, 7]
```

## Plot

Just pass the data in the `plot` function and it'll generate a plot, then call the `show` function to see the plot

```python
plt.plot(sunny_days)
plt.show()
```

## Change x-axis data

By default python uses the position of each item in the list as x-value, let's add month names instead

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
```

```python
plt.plot(months, sunny_days)
plt.show()
```

## Add title, labels and data point markers

Some commonly used markers are:  
 'o' - circle  
 's' - square  
 '^' - triangle  
 '*' - asterisk/star

```python
plt.plot(months, sunny_days, marker = "s")
plt.title("Sunny Days by month", fontsize = 20)
plt.xlabel("Month", fontsize = 14)
plt.ylabel("Sunny Days", fontsize = 14)
plt.show()
```

## Plot horizontal line

We can add a horizontal line showing the average number of sunny days between all the months. This can be done using the `axhline` function.

```python
avg_sunny_days = sum(sunny_days)/len(sunny_days)
```

```python
plt.plot(months, sunny_days, marker = "s")
plt.title("Sunny Days by month", fontsize = 20)
plt.xlabel("Month", fontsize = 14)
plt.ylabel("Sunny Days", fontsize = 14)
plt.axhline(avg_sunny_days)
plt.show()
```

Everything looks the same, let's add some style

## Use line style/colors

We can add linestyle and color to the horizontal line for average by using following arguments in the `axhline` function:  
`linestyle`: customizes line style (e.g., '--', '-.', ':').  
`color`: customizes line color (e.g., 'orange', 'red', 'gray').  

```python
plt.plot(months, sunny_days, marker = "s")
plt.title("Sunny Days by month", fontsize = 20)
plt.xlabel("Month", fontsize = 14)
plt.ylabel("Sunny Days", fontsize = 14)
plt.axhline(avg_sunny_days, linestyle = "-.", color = "orange")
plt.show()
```

## Add a legend

As a final touch, we can add legend to define what each line represent in the plot.  
We can use the `legend` function for adding a legend, but first we need to add `label` in `plot` and `axhline` and then calling the legend function will show those labels on the plot.

```python
plt.plot(months, sunny_days, marker = "s", label = "Sunny Days")
plt.title("Sunny Days by month", fontsize = 20)
plt.xlabel("Month", fontsize = 14)
plt.ylabel("Sunny Days", fontsize = 14)
plt.axhline(avg_sunny_days, linestyle = "-.", color = "orange", label = "Average")
plt.legend()
plt.show()
```
